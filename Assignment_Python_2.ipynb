```python
# 1. Write a function to find the second largest element in a list.
def second_largest(lst):
    unique = list(set(lst))
    unique.sort()
    return unique[-2] if len(unique) >= 2 else None

print(second_largest([10, 20, 4, 45, 45, 20]))
```

```python
# 3. Write a program to check if two lists are rotations of each other.
def is_rotation(lst1, lst2):
    return len(lst1) == len(lst2) and ''.join(map(str, lst2)) in ''.join(map(str, lst1*2))

print(is_rotation([1,2,3,4], [3,4,1,2]))
```

```python
# 4. Create a function that returns the intersection of two lists without using set().
def intersection(lst1, lst2):
    return [x for x in lst1 if x in lst2]

print(intersection([1,2,3,4], [3,4,5,6]))
```

```python
# 5. Write a program to flatten a nested list using recursion.
def flatten(lst):
    res = []
    for i in lst:
        if isinstance(i, list):
            res.extend(flatten(i))
        else:
            res.append(i)
    return res

print(flatten([1,[2,[3,4]],5]))
```

```python
# 8. Create a function that groups list elements based on their length.
def group_by_length(lst):
    res = {}
    for word in lst:
        res.setdefault(len(word), []).append(word)
    return res

print(group_by_length(["cat","dog","mouse","rat","bat"]))
```

```python
# 9. Write a program to find all pairs in a list whose sum equals a given number.
def pairs_sum(lst, target):
    res = []
    for i in range(len(lst)):
        for j in range(i+1, len(lst)):
            if lst[i] + lst[j] == target:
                res.append((lst[i], lst[j]))
    return res

print(pairs_sum([1,2,3,4,5], 5))
```

```python
# 11. Write a generator that yields prime numbers up to n.
def primes_gen(n):
    for num in range(2, n+1):
        for i in range(2, int(num**0.5)+1):
            if num % i == 0:
                break
        else:
            yield num

print(list(primes_gen(20)))
```

```python
# 12. Write a program to merge two dictionaries and sum values of common keys.
d1 = {'a': 10, 'b': 5}
d2 = {'a': 3, 'c': 7}
merged = {k: d1.get(k,0)+d2.get(k,0) for k in set(d1)|set(d2)}
print(merged)
```

```python
# 13. Implement binary search using a function.
def binary_search(lst, target):
    lst.sort()
    low, high = 0, len(lst)-1
    while low <= high:
        mid = (low+high)//2
        if lst[mid] == target:
            return mid
        elif lst[mid] < target:
            low = mid+1
        else:
            high = mid-1
    return -1

print(binary_search([1,3,5,7,9], 5))
```

```python
# 15. Write a program to validate a password using rules (length, digit, special char).
import re

def validate_password(pwd):
    return bool(re.match(r'^(?=.*[0-9])(?=.*[!@#$%^&*])[A-Za-z0-9!@#$%^&*]{8,}$', pwd))

print(validate_password("Passw0rd!"))
```

```python
# 17. Write a program to check if a string has balanced parentheses.
def balanced_parentheses(s):
    stack = []
    for char in s:
        if char == '(':
            stack.append(char)
        elif char == ')':
            if not stack: return False
            stack.pop()
    return not stack

print(balanced_parentheses("(a+b)*(c+d)"))
```

```python
# 18. Write a function that returns cumulative sum of a list.
def cumulative_sum(lst):
    res = []
    total = 0
    for x in lst:
        total += x
        res.append(total)
    return res

print(cumulative_sum([1,2,3,4]))
```

```python
# 20. Create a class with methods to add, remove, and display items.
class ItemList:
    def __init__(self):
        self.items = []
    def add(self, item):
        self.items.append(item)
    def remove(self, item):
        if item in self.items:
            self.items.remove(item)
    def display(self):
        print(self.items)

lst = ItemList()
lst.add("apple")
lst.add("banana")
lst.display()
lst.remove("apple")
lst.display()
```
