```python
# What is the distribution of 'Batting Averages'? Is it normally distributed or skewed?
sns.histplot(df['Batting Average'], kde=True)
plt.title("Distribution of Batting Averages")
plt.show()

# slightly right skewed distribution with a long tail on the right side, indicating that while most players have moderate batting averages, there are a few players with exceptionally high averages.
```

```python
# Are there any outliers in 'Strike Rate'? (e.g., players scoring exceptionally fast or slow).
sns.boxplot(x=df['Strike Rate'])
plt.title("Outliers in Strike Rate")
plt.show()
# Answer: Points beyond the whiskers are outliers (exceptionally fast or slow scorers).
```

```python
# What is the spread of 'Sixes' hit by the players?
sns.boxplot(x=df['Sixes'])
plt.title("Spread of Sixes Hit")
plt.show()
```

```python

# Bivariate Analysis (Relationships & Comparisons)



# Runs vs. Matches: Is there a strong linear correlation between the number of matches played and total runs scored?
sns.scatterplot(data=df, x='Matches', y='Runs')
sns.regplot(data=df, x='Matches', y='Runs', scatter=False)
plt.show()

df[['Matches','Runs']].corr()
```

```python
# Strike Rate vs. Average: Do players with high strike rates tend to have lower batting averages?
sns.scatterplot(data=df, x='Strike Rate', y='Batting Average')
sns.regplot(data=df, x='Strike Rate', y='Batting Average', scatter=False)
plt.show()

df[['Strike Rate','Batting Average']].corr()
```

```python
# Country vs. Performance: Which country has the highest average Batting Average across its players?
df.groupby('Country')['Batting Average'].mean().sort_values(ascending=False)
```

```python
sns.barplot(data=df, x='Country', y='Batting Average', estimator='mean')
plt.show()
```

```python
# Boundaries vs. Runs: What is the relationship between the number of 'Fours' and Total 'Runs'?
sns.scatterplot(data=df, x='Fours', y='Runs')
sns.regplot(data=df, x='Fours', y='Runs', scatter=False)
plt.show()

df[['Fours','Runs']].corr()
```

```python
# Conversion Rate: Which player has the best conversion rate of Fifties to Centuries? (i.e., Who capitalizes most on a good start?)
```

```python
df['Conversion Rate'] = df['Centuries'] / (df['Centuries'] + df['Fifties'])
df.loc[df['Conversion Rate'].idxmax(), ['Player Name', 'Conversion Rate']]
```

```python

sns.barplot(data=df.sort_values('Conversion Rate', ascending=False).head(10),
            x='Player Name', y='Conversion Rate')
plt.show()

```

```python
# Not Outs vs. Average: Does having a high number of 'Not Outs' significantly inflate a player's Batting Average?
sns.scatterplot(data=df, x='Not Outs', y='Batting Average')
sns.regplot(data=df, x='Not Outs', y='Batting Average', scatter=False)
plt.show()

df[['Not Outs','Batting Average']].corr()
```

```python
# Balls Faced vs. Sixes: Do players who face more balls hit significantly more sixes, or is it a trait of specific aggressive players?
sns.scatterplot(data=df, x='Balls Faced', y='Sixes')
sns.regplot(data=df, x='Balls Faced', y='Sixes', scatter=False)
plt.show()

df[['Balls Faced','Sixes']].corr()
```
