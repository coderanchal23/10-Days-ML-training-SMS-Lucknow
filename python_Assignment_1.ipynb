```python
# 21. Write a recursive function to compute the power of a number.
def power(a, b):
    if b == 0:
        return 1
    return a * power(a, b-1)

print(power(2, 4))
```

```python
# 22. Write a function that returns all prime numbers up to n.
def primes(n):
    res = []
    for num in range(2, n+1):
        for i in range(2, int(num**0.5)+1):
            if num % i == 0:
                break
        else:
            res.append(num)
    return res

print(primes(20))
```

```python
# 23. Write a recursive function to flatten a nested list.
def flatten(lst):
    res = []
    for i in lst:
        if isinstance(i, list):
            res.extend(flatten(i))
        else:
            res.append(i)
    return res

print(flatten([1,[2,[3,4]],5]))
```

```python
# 24. Create a function that accepts a list and returns a dictionary of element frequencies.
def freq(lst):
    d = {}
    for i in lst:
        d[i] = d.get(i, 0) + 1
    return d

print(freq([1,2,2,3,3,3]))
```

```python
# 25. Write a function that accepts another function as an argument and applies it to each element of a list.
def apply_func(func, lst):
    return [func(x) for x in lst]

print(apply_func(lambda x: x**2, [1,2,3,4]))
```

```python
# 27. Compress a string using character counts (e.g., "aaabb" â†’ "a3b2").
def compress(s):
    res = ""
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            count += 1
        else:
            res += s[i-1] + str(count)
            count = 1
    res += s[-1] + str(count)
    return res

print(compress("aaabb"))
```

```python
# 28. Find all sublists of a list whose sum is zero.
def zero_sum_sublists(lst):
    res = []
    n = len(lst)
    for i in range(n):
        total = 0
        for j in range(i, n):
            total += lst[j]
            if total == 0:
                res.append(lst[i:j+1])
    return res

print(zero_sum_sublists([3,4,-7,1,3,-4,-2,-2]))
```
